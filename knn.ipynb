```python
import math
import random
import operator
import numpy as np
import pandas as pd

def euclidian_distance(instance1, instance2):
    distance = 0
    distance += pow(instance1 - instance2, 2)
    return math.sqrt(distance)

def manhattan_distance(instance1 ,instance2):
    distance = 0
    distance += (abs(instance1 - instance2))
    return distance    

def hamming_distance(instance1, instance2):
    var_sim = 0
    if (instance1 == instance2):
        var_sim = 0
    else:
        var_sim = 1
    return var_sim    
        
def getNeighbors(trainSet, testInstance, k, type_features):
    distances = []
    
    length = len(testInstance)
    for x in range(len(trainSet)):
        dist_arr = []
        for y in range(len(trainSet[x])):
            if(type_features[y] == 'n'):
                dist = euclidian_distance(testInstance[y], trainSet[x][y])
            else:
                dist = hamming_distance(testInstance[y], trainSet[x][y])
            dist_arr.append(dist)
            ans = np.sum(np.array(dist)**2,axis=0)**.5
        distances.append((trainSet[x], ans))
    distances.sort(key=operator.itemgetter(1))
    
    neighbors = []
    for x in range(k):
        print(distances[x][0])
        neighbors.append(distances[x][0])
    return neighbors
    #return dist_arr

def getResponse(neighbors, label):
    classVotes = {}
    for x in range(len(label)):
        response = label[x]
        print("Response", response)
        if response in classVotes:
            classVotes[response] += 1
        else:
            classVotes[response] = 1
    print(classVotes.items())        
    #sortedVotes = sorted(classVotes.items(), key=operator.itemgetter(1), reverse="True")
    sortedVotes = sorted(classVotes.items(), key = lambda x: x[1], reverse=True)
    return sortedVotes[0][0]

# def getResponse(neighbors):
#     classVotes = {}
#     for x in range(len(neighbors)):
#         response = neighbors[x][-1]
#         print("Response", response)
#         if response in classVotes:
#             classVotes[response] += 1
#         else:
#             classVotes[response] = 1
#     print(classVotes.items())        
#     sortedVotes = sorted(classVotes.items(), key=operator.itemgetter(1), reverse=True)
#     return sortedVotes[0][0]        
```
